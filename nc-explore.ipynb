This part of the exercise is rather disorienting if you do not have a familarity with the map projection metadata this file is making use of. Fortunately, we are given a hint by the `GRIB-1 GDS data representation type` description  of the `grid_type_code` variable.

A simple Google search of that description takes us to [[http://www.wmo.int/pages/prog/www/WMOCodes/Guides/GRIB/GRIB1-Contents.html][A GUIDE TO THE CODE FORM FM 92-IX Ext. GRIB Edition 1 from 1994]], Therein one can find an explanation of the variables needed to understand the map projection. Let's review these. We will need to import netcdf4-python first and open the file with that API.


```python
import netCDF4
f = netCDF4.Dataset('data.nc')

grid_type_code = f.variables['grid_type_code'][0]
print 'grid type code is ' + str(grid_type_code)
```

A `grid_type_code` of `5` corresponds to a projection of `Polar Stereographic`. (See table 6.) 

```python
grid_type = f.variables['grid_type']
print 'The grid type is ' + ''.join(grid_type[0,:])
```

Note that this newest piece of information relating to a Lambert Conformal Projection stands in contrast to the earlier projection information about a Polar Stereographic projection. There is a bug in the metadata description of the projection, but let's move on anyway. Next let's grab the number of points along the X and Y axes as given by `Nx` and `Ny`.

```python
nx, ny = f.variables['Nx'][0], f.variables['Ny'][0]
print nx, ny
```

Next the `first latitude`, and `first longitude` (whatever that means).

```python
la1, lo1 = f.variables['La1'][0], f.variables['Lo1'][0]
# print la1[0], la1.units, lo1[0], lo1.units
print la1, lo1
```
